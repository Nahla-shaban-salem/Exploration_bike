## Univariate Exploration



```python
bike.info()
```

```python
#Change datatype for date
bike['start_time'] = pd.to_datetime(bike['start_time'], errors='coerce');
bike['end_time'] = pd.to_datetime(bike['end_time'], errors='coerce');
bike.info();
```

```python
#Knowing nan values
bike.isna().sum()
```

```python
#Represent nan value in graph
na_counts = bike.drop(['duration_sec', 'start_time', 'end_time', 'start_station_latitude',
                             'start_station_longitude','end_station_latitude','end_station_longitude','bike_id','user_type'
                            ,'bike_share_for_all_trip'], axis=1).isna().sum()
print(na_counts)
sb.barplot(na_counts.index.values, na_counts,color ='violet')
plt.xticks(rotation=90) 
for i in range (na_counts.shape[0]):
    count = na_counts[i]
    plt.text(i, count+400, count, ha = 'center', va='top')
```

## Bivariate Exploration



```python
bike.head(1)
```

```python
#from start time extract days and use it to know traffic in days
bike['Days'] = bike["start_time"].dt.day.astype('int')

#sb.pointplot(data=bike ,)
plt.figure(figsize=(10,7));
bike.groupby('Days')['start_time'].count().plot(kind='line',title='Count rides by days');

plt.xticks(rotation=15);
plt.xlabel('Days')
plt.ylabel('Count')
;
```

```python
#from member_birth_year extract age for members
bike['age'] = 2020-bike['member_birth_year'].astype('int')
sb.boxplot(data=bike, x='age', color='red');
plt.xlabel('Age');
plt.title('Age box plot');
#most of age between 28 to 40 and mean 33 and there are some outlier
```

```python
sb.boxplot(data=bike, x='age',y='member_gender', color='firebrick');
plt.xlabel('Age');
plt.ylabel('Gender');
plt.title('Age VS Gender box plot');
```

```python
#to know top 10 start_station_name
bike['start_station_name'].value_counts()
```

```python
#extract top 10 start_station_name and use it
bike2=bike.query('start_station_name in["Market St at 10th St","San Francisco Caltrain Station 2  (Townsend St at 4th St)","Berry St at 4th St","Montgomery St BART Station (Market St at 2nd St)","Powell St BART Station (Market St at 4th St)","San Francisco Caltrain (Townsend St at 4th St)","San Francisco Ferry Building (Harry Bridges Plaza)","Howard St at Beale St","Steuart St at Market St","Powell St BART Station (Market St at 5th St)"]')
plt.figure(figsize=(8,4));
sb.boxplot(data=bike2, y='start_station_name', x='age',color='maroon');
plt.xlabel('Age');
plt.ylabel('start_station_name');
plt.title('start_station_name VS Age');
```

```python
#for top 10 which Customer or Subscriber
plt.figure(figsize=(8,4));

sb.countplot(data = bike2, x = 'start_station_name', hue = 'user_type')
plt.xticks(rotation=90);

plt.title('user_type,start_station_name by count');
# from Graph most of members are subscriber
```
